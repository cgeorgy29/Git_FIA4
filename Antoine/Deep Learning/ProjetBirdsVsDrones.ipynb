```python
import os
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

nrows = 4
ncols = 4
i = 0

pic_index = 0

fig = plt.gcf()
fig.set_size_inches(ncols*4, nrows*4)

pic_index+=8
bird_images = os.listdir('C:/Users/avignero/Desktop/bird_vs_drone/bird/training/')
next_bird_pix = [os.path.join('C:/Users/avignero/Desktop/bird_vs_drone/bird/training/', fname) 
                for fname in bird_images[pic_index-8:pic_index] 
               ]

print(len(next_bird_pix))

for i, img_path in enumerate(next_bird_pix):
    #print(item)
    sp = plt.subplot(nrows, ncols, i + 1)
    sp.axis('On')
    img = mpimg.imread(img_path)
    plt.imshow(img)

plt.show()
```
